```python
def calSameNum(pd_dia):
    pd_same4 = pd_dia[(pd_dia["admission_situation"]==pd_dia["treat_process"]) & (pd_dia["admission_situation"] ==pd_dia["discharge_situation"] ) &(pd_dia["admission_situation"] ==pd_dia["discharge_order"])]
    print(pd_same4.shape)
    
    pd_same3 = pd_dia[(pd_dia["admission_situation"]==pd_dia["treat_process"]) & (pd_dia["admission_situation"] ==pd_dia["discharge_situation"] )]
    print(pd_same3.shape)
        
    pd_same2 = pd_dia[(pd_dia["admission_situation"]==pd_dia["treat_process"])]
    print(pd_same2.shape)
calSameNum(pd_dia)
```

```python
def calNONum(pd_dia):
    pd_no1= pd_dia[(pd_dia["admission_situation"]!=pd_dia["treat_process"]) ]
    print(pd_no1.shape)
    
    pd_no2 = pd_dia[(pd_dia["admission_situation"]!=pd_dia["discharge_situation"])]
    print(pd_no2.shape)
        
    pd_no3 = pd_dia[(pd_dia["admission_situation"]!=pd_dia["discharge_order"])]
    print(pd_no3.shape)
    
    pd_no4 = pd_dia[(pd_dia["admission_situation"]!=pd_dia["treat_process"])&((pd_dia["admission_situation"]!=pd_dia["discharge_order"])) ]
    print(pd_no4.shape)
calNONum(pd_dia)
```
